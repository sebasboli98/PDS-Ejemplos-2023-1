## * Demostración de la conversión de señal analógica a digital: Muestreo, Quantización y Gráfico de comparación

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear señal analógica
t = np.linspace(0, 1, num=1000)
f = 1
s = np.sin(2*np.pi*f*t)

# Muestreo de la señal
fs = 10 # frecuencia de muestreo
t_muestreo = np.linspace(0, 1, num=fs)
s_muestreo = np.sin(2*np.pi*f*t_muestreo)

# Quantización de la señal
bits = 8 # número de bits
L = 2**bits # niveles de quantización
delta = 2/L # tamaño del cuantizador
s_quant = np.round(s_muestreo/delta)*delta # señal cuantizada

# Gráfico de la señal analógica y cuantizada
plt.plot(t, s, label='Analógica')
plt.stem(t_muestreo, s_quant, label='Digital', use_line_collection=True)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

## * Demostración grafica de cuando no cumple el teorema de Nyquist (aliasing presente) y conversión de señal analógica a digital con muestreo y quantización"

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear señal analógica
t = np.linspace(0, 1, num=1000)
f = 2
s = np.sin(2*np.pi*f*t)

# Muestreo de la señal
fs = 5 # frecuencia de muestreo
t_muestreo = np.linspace(0, 1, num=fs)
s_muestreo = np.sin(2*np.pi*f*t_muestreo)

# Verificación de la frecuencia de Nyquist
nyquist_frequency = fs/2
if f > nyquist_frequency:
    print("La frecuencia de la señal es mayor que la frecuencia de Nyquist.")

# Quantización de la señal
bits = 3 # número de bits
L = 2**bits # niveles de quantización
delta = 2/L # tamaño del cuantizador
s_quant = np.round(s_muestreo/delta)*delta # señal cuantizada

# Gráfico de la señal analógica y cuantizada
plt.plot(t, s, label='Analógica')
plt.stem(t_muestreo, s_quant, label='Digital', use_line_collection=True)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

## * Generación y representación gráfica de una señal sinusoidal digitalizada mediante muestreo y quantización. Se garantiza que no hay aliasing

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_senal(f, fs, bits):
    # Crear señal analógica
    t = np.linspace(0, 1, num=1000)
    s = np.sin(2*np.pi*f*t)

    # Muestreo de la señal
    t_muestreo = np.linspace(0, 1, num=fs)
    s_muestreo = np.sin(2*np.pi*f*t_muestreo)

    # Quantización de la señal
    L = 2**bits # niveles de quantización
    delta = 2/L # tamaño del cuantizador
    s_quant = np.round(s_muestreo/delta)*delta # señal cuantizada

    # Gráfico de la señal analógica y cuantizada
    plt.plot(t, s, label='Analógica')
    plt.stem(t_muestreo, s_quant, label='Digital', use_line_collection=True)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.show()

if __name__ == '__main__':
    # Solicitar frecuencia de la señal y frecuencia de muestreo al usuario
    f = int(input("Ingrese la frecuencia de la señal (en Hz): "))
    fs = int(input("Ingrese la frecuencia de muestreo (en Hz): "))
    while 2*f >= fs:
        print("La frecuencia de muestreo debe ser mayor al doble de la frecuencia de la señal para evitar el aliasing.")
        fs = int(input("Ingrese una frecuencia de muestreo mayor que 2*f: "))
    bits = int(input("Ingrese el número de bits para la quantización: "))

    # Generar y mostrar la señal
    generar_senal(f, fs, bits)
```
