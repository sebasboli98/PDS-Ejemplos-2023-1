```python
import numpy as np
import matplotlib.pyplot as plt

# Define la frecuencia de muestreo y la duración de la señal
fs = 1000 # Hz
duration = 5 # segundos

# Genera la señal de rampa
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
signal = t

# Digitaliza la señal mediante la técnica de muestreo
num_samples = t.shape[0]
sampled_signal = signal[::int(num_samples/10)]
sampled_time = t[::int(num_samples/10)]

# Grafica la señal original y la señal digitalizada
plt.figure(figsize=(12, 6))
plt.plot(t, signal, label='Señal original')
plt.stem(sampled_time, sampled_signal, label='Señal digitalizada', markerfmt='x')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```
