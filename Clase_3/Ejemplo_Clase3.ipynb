```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la señal rampa
N = 1000 # número de muestras de la señal
rampa = np.linspace(0, 1, N) # señal rampa de 0 a 1
f_s = 100 # frecuencia de muestreo
t = np.arange(N) / f_s # tiempo discreto

# Graficamos la señal rampa
plt.plot(t, rampa, marker='o')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal rampa digital')
plt.show()

# Calculamos la energía de la señal rampa
energia = np.sum(rampa**2) / f_s
print('La energía de la señal rampa es:', energia)

# Calculamos la potencia de la señal rampa
potencia = np.mean(rampa**2)
print('La potencia de la señal rampa es:', potencia)
```
