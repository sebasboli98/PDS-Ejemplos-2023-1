```python
import numpy as np
import matplotlib.pyplot as plt

# Crear señal de entrada con dos componentes sinusoidales
fs = 1000  # Frecuencia de muestreo en Hz
f1 = 10  # Frecuencia de la primera componente sinusoidal en Hz
f2 = 100  # Frecuencia de la segunda componente sinusoidal en Hz
A1 = 1  # Amplitud de la primera componente sinusoidal
A2 = 0.5  # Amplitud de la segunda componente sinusoidal
t = np.arange(0, 1, 1/fs)  # Vector de tiempo en segundos
x = A1*np.sin(2*np.pi*f1*t) + A2*np.sin(2*np.pi*f2*t)  # Señal de entrada

# Aplicar filtro de media móvil con longitud 20
N = 20  # Longitud del filtro
b = np.ones(N) / N  # Coeficientes del filtro
y = np.convolve(x, b, mode='same')  # Señal filtrada

# Graficar señal de entrada y señal filtrada en subplots separados
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
fig.subplots_adjust(hspace=0.5)

# Subplot 1: señal de entrada
ax1.plot(t, x)
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Amplitud')
ax1.set_title('Señal de entrada')

# Subplot 2: señal filtrada
ax2.plot(t, y)
ax2.set_xlabel('Tiempo (s)')
ax2.set_ylabel('Amplitud')
ax2.set_title('Señal filtrada con filtro de media móvil de longitud 20')

plt.show()
```
