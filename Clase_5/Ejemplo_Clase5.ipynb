### El c칩digo utiliza la convoluci칩n para encontrar la respuesta de un sistema ante un impulso dado. Primero, se define una se침al de impulso y una respuesta al impulso del sistema. Luego, se convoluciona el impulso con la respuesta al impulso para obtener la respuesta al impulso del sistema

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la respuesta del sistema
h = np.array([1, 2, 1])



delta = np.zeros(len(h))
delta[0] = 1




# Calcular la respuesta al impulso utilizando la convoluci칩n
y = np.convolve(delta, h, mode='full')

# Calcular la salida del sistema ante un impulso
x = np.zeros(100)
for i in range(len(delta)):
    x[i] = np.sum(delta[:i+1] * h[:i+1])

# Graficar la respuesta al impulso del sistema
plt.figure(figsize=(8, 6))
plt.stem(range(len(y)), y)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Respuesta al impulso del sistema')
plt.show()

# Graficar la salida del sistema ante un impulso
plt.figure(figsize=(8, 6))
plt.stem(range(len(x)), x)
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.title('Salida del sistema ante un impulso')
plt.show()
```
