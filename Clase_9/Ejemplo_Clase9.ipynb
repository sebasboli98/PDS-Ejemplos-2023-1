```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Definimos la función de transferencia del sistema discreto de tiempo LTI
numerator = [1, -1.5, 0.7]
denominator = [1, -1.3, 0.5]
H = signal.TransferFunction(numerator, denominator, dt=1)

# Calculamos la respuesta al impulso del sistema
impulse_response = signal.dimpulse(H)

# Extraemos la secuencia de salida de la respuesta al impulso
h = impulse_response[1][0]

# Definimos la variable simbólica z
z = signal.zpk2tf([], [0.5, 0.7], 1)

# Calculamos la transformada inversa Z de la secuencia de salida para encontrar la respuesta al impulso en el dominio del tiempo discreto
t, h_time_domain = signal.dlti(*z).output(h, np.arange(len(h)))

# Graficamos la respuesta al impulso del sistema en el dominio del tiempo discreto
plt.plot(t, h_time_domain)
plt.title('Respuesta al impulso en el dominio del tiempo discreto')
plt.xlabel('Tiempo discreto')
plt.ylabel('Amplitud')
plt.show()
```
